```python
import re
import pandas as pd
```

```python
with open('/content/bukvari.txt', 'r', encoding='utf-8') as file:
    text = file.read()

print(text[:500])  # Вывод первых 500 символов
```

```python
clean_text = re.sub(r"[^\w\s\-–]+|\d+", "", text)

print (clean_text[:500])
```

```python
# Сохранение очищенного текста в новый файл
with open('/content/cleaned_text_2.txt', 'w', encoding='utf-8') as file:
    file.write(clean_text)
```

```python
# Извлечение слов с дефисами или тире
hyphenated_words = re.findall(r"\b\w{1,4}(?:[-–]\w{1,4})+\b", clean_text) # с учетом количества

# Создание DataFrame
df = pd.DataFrame({"Hyphenated_Words": hyphenated_words})
```

```python
#файл, почищенный gpt: все равно остаются сочетания букв
with open('/content/filtered_russian_syllabic_words.csv', 'r', encoding='utf-8') as file:
    df_filtered = pd.read_csv(file)
df_filtered
```

```python
# Убираем сочетания заглавных букв, так как онивсегда или почти всегда не несут смысла

# Функция для исключения слов, где все буквы заглавные и есть дефис или тире
def is_not_all_uppercase_with_hyphen(word):
    return not bool(re.fullmatch(r"[А-ЯЁ]+(?:[-–][А-ЯЁ]+)+", word))

# Фильтруем слова
filtered_lowercase_words = df_filtered[df_filtered["Hyphenated_Words"].apply(is_not_all_uppercase_with_hyphen)]

# Сохраняем результат в новый CSV-файл
filtered_lowercase_words.to_csv("filtered_lowercase_words.csv", index=False, encoding="utf-8")
```

```python
filtered_lowercase_words
```

```python
# Приводим у нижнему регистру и убираем повторы

# Приводим все слова в столбце "Hyphenated_Words" к нижнему регистру
filtered_lowercase_words.loc[:, "Hyphenated_Words"] = filtered_lowercase_words["Hyphenated_Words"].str.lower()

# Удаляем дубликаты
filtered_hyphenated_words_unique = filtered_lowercase_words.drop_duplicates(subset=["Hyphenated_Words"])

# Сохраняем результат в новый CSV-файл
filtered_hyphenated_words_unique.to_csv("filtered_hyphenated_words_unique.csv", index=False, encoding="utf-8")
```

```python
# Вручную удалили двух-трех, фе-фё-ла, ам-ам, му-му, ко-ко-ко, тук-тук-тук, пи-пи-пи, пых-пых, кну-кну, кно-кно,
# тка-тка, га-га, го-гу, га-гу, га-га-га, бим-бом, бум-бум, бам-бам, он-тан (фонтан был), ди-ди-ти, пам-бам-пам, бум-бум-пум
# цып-цып, цок-цок, как-то, чё-гко и разделили "гла-за фо-на-ря"

# Открываем для просмотра
with open('/content/bukvari_cleaned.csv', 'r', encoding='windows-1251') as file:
    df_bukvari = pd.read_csv(file)

# Просмотр первых строк
print (df_bukvari[:20])
```

```python
df_bukvari
```

```python
# Теперь создаем копию первого столбца в df_bukvari со словами без дефисов

# Создаем второй столбец как копию первого
df_bukvari["Words"] = df_bukvari["Hyphenated_Words"]

# Удаляем дефисы в новом столбце
df_bukvari["Words"] = df_bukvari["Words"].replace("-", "", regex=True)

df_bukvari
```

```python
# Создаем словарь, где ключи - слова из второго столбца, а значения - из первого
words_dict = dict(zip(df_bukvari["Words"], df_bukvari["Hyphenated_Words"]))

# Проверка результата
print(words_dict)
```

```python
import json

# Сохраняем словарь в JSON файл
with open("hyphenated_words_dict.json", "w", encoding="utf-8") as json_file:
    json.dump(words_dict, json_file, ensure_ascii=False, indent=4)
```
